```python
import  pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/digipodium/Datasets/main/regression/bike_rental.csv')
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error ,r2_score
```

```python
df.columns
```

```python
X=df.drop(['cnt','casual','registered','dteday'],axis=1)
y=df['cnt']
```

```python
Xtrain,ytrain,Xtest,ytest=train_test_split(X,y,test_size=0.2)
```

```python
reg=xgb.XGBRegressor(n_estimators=100,learning_rate=.1)
reg.fit(Xtrain,ytrain)
ypred=reg.predict(Xtrain)
ypredtest=reg.predict(Xtest)
print(f'r2{r2_score(ytest,ypredtest)}')
print(f'r2 train{r2_score(ytrain,ypred)}')
```
