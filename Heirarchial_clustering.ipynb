```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt

```

```python
pd.options.plotting.backend='plotly'
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/digipodium/Datasets/main/classfication/Social_Network_Ads.csv")
```

```python
df.head()
```

```python
df.plot(x="Age",y='EstimatedSalary',kind='scatter',)
```

```python
from scipy.cluster.hierarchy import dendrogram ,linkage
from sklearn.preprocessing import StandardScaler
```

```python
X=df[['Age','EstimatedSalary']]
ss=StandardScaler()
```

```python
ss.fit_transform(X)
```

```python
fig,ax=plt.subplots(figsize=(15,10))
Z=linkage(X,'ward')
dendrogram(Z,ax=ax)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering 
```

```python
df['group']=AgglomerativeClustering(n_clusters=2).fit_predict(X)
df.plot(kind='scatter',x='Age',y='EstimatedSalary',color='group',)
```

```python
df['groups']=AgglomerativeClustering(n_clusters=3).fit_predict(X)
df.plot(kind='scatter',x='Age',y='EstimatedSalary')
```

```python
px.density_contour(
    df,
    x='Age',
    y='EstimatedSalary',
    color='groups'
)
```
